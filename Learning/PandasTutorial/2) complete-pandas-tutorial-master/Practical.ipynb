```python
coffee.loc[(3,4), 'Units Sold'] = np.nan
```

```python
coffee.isna().sum() # Counts the number of NAN values
```

```python
coffee = coffee.fillna(10000) # replaces the na values with any given number

coffee= coffee.fillna(coffee['Units Sold'].interpolate()) # This interpolates the gap between the top and bottom

coffee['Units Sold'] = coffee['Units Sold'].interplate() # interpolate over the whole thing 

coffee.dropna() # Drops all things with *any* nas - be careful here.

coffee[coffee['Units Sold'].isna()] # returns the rows with nas in them
coffee[coffee['Units Sold'].notna()] # returns the rows with no nas in them
```

```python
coffee['Units Sold'] = coffee['Units Sold'].interpolate() # Interpolates over the units sold

coffee.groupby(['Coffee Type'])['Units Sold'].sum() # Gives the sum total of all coffees sold across type

coffee.groupby(['Coffee Type'])['Units Sold'].mean() # Gives the average of all coffees sold across type on any day.

coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum', 'price': "mean"})
# Gives the sum of all coffees sold, and the average price of their sale

coffee.groupby(['Coffee Type',"Day"]).agg({'Units Sold': 'sum', 'price': "mean"})
# Gives the sum of all coffees sold, and the average price of their sale, grouped by day
```

```python
pivot = coffee.pivot(columns = "Coffee Type", index="Day", values="price")

#can use a pivot to take a grouping structure of the day that looks at the price for any given day

pivot
```

```python
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)

#creates a new coffee variable that looks at yesterday's revenue using the shift function (recursion value)
# now 2 because we are looking past the Late/Espresso setup 

coffee['pct change'] = coffee['revenue']/coffee['revenue'].shift(2)

#We have a percentage change value associated with the series
#Be sure to use github copilot if you have it installed! It's about making yourself better! (Ctrl + I)
```

```python
coffee.select_dtypes('float').cumsum()

# operates a cumulative sum based on all float data types
```

```python
coffee['cum_rev'] = coffee['revenue'].cumsum()

# operates a cumulative sumthat accumulates all the previous data series
```

```python
latte = coffee[coffee['Coffee Type'] == "Latte"].copy()

#selecting only the latte data type

latte['3day'] = latte['Units Sold'].rolling(3).sum

# operates a cumulative sumthat accumulates over the past three observations
```
