We then build our data importing functions. Draws on macrobond_api_constants, we define a frequency as f, and we make it monthly.

Then we define a get data function, with defined tickers and column names for each.

This then pulls the data for a specified frequency - in the cases of daily, monthly, quarterly and annually.

```python
from macrobond_api_constants import SeriesFrequency as f
def get_data_d(tickers, column_names):
    ticks_new = [x.lower() for x in tickers]
    mb = m_macrobond.Macrobond()
    dat = mb.CreateUnifiedSeriesRequst(ticker_list=ticks_new, Frequency=f.DAILY)
    dat = dat.asfreq(freq='D')
    dat.columns = column_names
    return dat
```

```python
from macrobond_api_constants import SeriesFrequency as f
def get_data_m(tickers, column_names):
    ticks_new = [x.lower() for x in tickers]
    mb = m_macrobond.Macrobond()
    dat = mb.CreateUnifiedSeriesRequst(ticker_list=ticks_new, Frequency=f.MONTHLY)
    dat = dat.asfreq(freq='M')
    dat.columns = column_names
    return dat
```

```python
from macrobond_api_constants import SeriesFrequency as f
def get_data_q(tickers, column_names):
    ticks_new = [x.lower() for x in tickers]
    mb = m_macrobond.Macrobond()
    dat = mb.CreateUnifiedSeriesRequst(ticker_list=ticks_new, Frequency=f.QUARTERLY)
    dat = dat.asfreq(freq='Q')
    dat.columns = column_names
    return dat
```

```python
from macrobond_api_constants import SeriesFrequency as f
def get_data_a(tickers, column_names):
    ticks_new = [x.lower() for x in tickers]
    mb = m_macrobond.Macrobond()
    dat = mb.CreateUnifiedSeriesRequst(ticker_list=ticks_new, Frequency=f.HIGHEST)
    dat = dat.asfreq(freq='A')
    dat.columns = column_names
    return dat
```
